#### 2. chr() dan ord()


```python

````python
integer = 78
print(integer)
print(type(integer))

ubah = chr(integer)
print(ubah)
print(type(ubah))

kembali = ord(ubah)
print(kembali)
print(type(kembali))
````

### Built-in Type Functions

#### 1. bytes()

```python

````python
y = bytes()
print(y)

x = bytes(5)
print(x)
print(type(x))

list = bytes([2, 3, 4, 6])
print(list)

angka_besar = bytes([3, 2, 1, 300])
print(angka_besar)
````

#### 2. str(), bool(), int(), float(), complex()

```python

````python
x = bool(1)
print(x)
````

````python
y = 5
print(str(y))
print(float(y))
````

### All Other Built-in Functions

#### 1. Len

````python
angka = [3, 4, 7, 11, 18, 29, 42]
print(f"panjang list adalah: {len(angka)}")
````

#### 2. min() dan max()

````python
list = [3, 4, 7, 11, 18, 29, 42]
print(f"Nilai maksimal adalah: {max(list)}")
print(f"Nilai minimal adalah: {min(list)}")
````

#### 3. sorted()

````python
a = ("h", "b", "a", "c", "f", "d", "e", "g")
x = sorted(a)
print(x)
````
